```python
import pandas as pd

# Load your data (this step depends on your file format)
file_path = '/Users/dadmehr/Codes/Python/Past/Call_Data_20240130.csv'
data = pd.read_csv(file_path)

# replacing '--' with nothing in the 'Final Call Type' column
data['Final Call Type'] = data['Final Call Type'].str.replace('--', '')
```

```python
# Now save the modified DataFrame back to a new CSV file
output_file_path = '/Users/dadmehr/Codes/Python/Past/modified_Call_Data_20240130.csv'
data.to_csv(file_path, index=False)

# 'index=False' is used to prevent pandas from writing row indices into the CSV file
```

```python
contains_2024 = data['Original Time Queued'].str.contains('2024', case=False, na=False)
data_2024 = data[contains_2024]
```

```python
import folium

# Initialize a map centered on UW
map = folium.Map(location=[47.655334, -122.303520], zoom_start=14) # Adjust zoom level as needed



data_2024_UW = data_2024[(data_2024['Blurred_Latitude'] < 47.675) &
                         (data_2024['Blurred_Latitude'] > 47.645) & 
                         (data_2024['Blurred_Longitude'] < -122.275) &
                         (data_2024['Blurred_Longitude'] > -122.32)]


# Add points to the map
for idx, row in data_2024_UW.iterrows():  # And here
    folium.CircleMarker(
        location=[row['Blurred_Latitude'], row['Blurred_Longitude']],
        radius=.3,  # Adjust the size of the circle markers
        popup=row['Final Call Type'],  # This can be any column in your DataFrame
        fill=True
    ).add_to(map)

# Display the map
map
```
