# TimeTree 

```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Initialize the WebDriver
driver = webdriver.Chrome()  # Or use 'webdriver.Firefox()' for Firefox

# Load the dataset from the CSV file
df = pd.read_csv("DNA data.csv")  # Replace "DNA_data.csv" with your dataset file name and path
df["Found Text"] = ""
# Limit the loop to the first 10 rows
for index, row in df.head(3).iterrows():
    data_name = row["Scientific Name"]
    human_name = row["Human"]
    human_name = "Homo sapiens"
    
    print("data_name = ", data_name)
    #print("human_name = ", human_name)
    

    # Initialize the WebDriver
    driver = webdriver.Chrome()  # Or use 'webdriver.Firefox()' for Firefox

    # Open the TimeTree webpage
    driver.get("http://timetree.org")

    # Find the input fields for the species names and enter the values
    species1_input = driver.find_element(By.ID, "taxon-a")
    species1_input.clear()
    species1_input.send_keys(human_name)

    species2_input = driver.find_element(By.ID, "taxon-b")
    species2_input.clear()
    species2_input.send_keys(data_name)

    # Find and click the search button
    search_button = driver.find_element(By.ID, "pairwise-search-button1")
    search_button.click()

    # Wait for the element containing the text "MYA" to be loaded
    try:
        # Adjust the timeout value as needed
        WebDriverWait(driver, 3).until(
            EC.presence_of_element_located(
                (By.XPATH, "//*[contains(text(), 'MYA')]")
            )
        )
        # Locate the element and extract its text
        mya_element = driver.find_element(
            By.XPATH, "//*[contains(text(), 'MYA)')]")
        mya_text = mya_element.text
        print(f"Found Text: {mya_text}")
        df.at[index, "Found Text"] = mya_text
    except TimeoutException:
        print("The element containing 'MYA' was not found within the given time.")
    finally:
        # Close the browser
        driver.quit()
```
